```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
iband_counts = Table.read("iband_mag_counts.dat", format='ascii')
```

```python
m_grid = iband_counts[iband_counts.colnames[0]]
Ntot = iband_counts[iband_counts.colnames[1:]]
```

```python
fig, ax = plt.subplots(1, figsize=(8,6))
for model in Ntot.colnames:
    ax.plot(m_grid, Ntot[model], label=model)
ax.set_yscale('log')
ax.legend()
plt.show()
```

```python
frac = Ntot[Ntot.colnames[1:]]
N_uniform = Ntot[Ntot.colnames[0]].data
for model in frac.colnames:
    frac[model] = frac[model]/N_uniform
```

```python
fig, ax = plt.subplots(1, figsize=(8,6))
for model in frac.colnames:
    ax.plot(m_grid, frac[model], label=model)
ax.legend()
plt.show()
```
